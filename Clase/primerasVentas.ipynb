1. Solicitar al usuario el monto de las primeras 10 ventas
2. Calcular el promedio de estas ventas iniciales
3. Identificar cuales ventas estuvieron por encima del promedio
4. Calcular el total recaudado en estas primeras ventas
5. Determinar cual fue tu mejor y peor venta inicial

```python
import numpy as np

montos_ventas = []
cantidad_ventas = 10

print("----------------- ANÁLISIS DE VENTAS ----------------\n")

# Solicitar al usuario el monto de las primeras 10 ventas
for i in range(cantidad_ventas):
    while True:
        try:
            monto = float(input(f"Ingrese el monto de la venta {i+1}: "))
            montos_ventas.append(monto)
            break
        except ValueError:
            print("Entrada inválida. Por favor, ingrese un número.")
print("\nLos montos de las ventas ingresadas son: ", montos_ventas)

# Convertir la lista a un array de numpy para usar sus funciones
ventas_array = np.array(montos_ventas)

# 2. Calcular el promedio de estas ventas iniciales
promedio_ventas = np.mean(ventas_array)
print(f"\n2. Promedio de las ventas: ${promedio_ventas:.2f}")

# 3. Identificar cuales ventas estuvieron por encima del promedio
ventas_sobre_promedio = ventas_array > promedio_ventas
indices_sobre_promedio = np.where(ventas_sobre_promedio)[0]
valores_sobre_promedio = ventas_array[ventas_sobre_promedio]

print(f"\n3. Ventas por encima del promedio:")
for i, indice in enumerate(indices_sobre_promedio):
    print(f"   Venta {indice + 1}: ${valores_sobre_promedio[i]:.2f}")

# 4. Calcular el total recaudado en estas primeras ventas
total_recaudado = np.sum(ventas_array)
print(f"\n4. Total recaudado: ${total_recaudado:.2f}")

# 5. Determinar cual fue la mejor y peor venta inicial
mejor_venta = np.max(ventas_array)
peor_venta = np.min(ventas_array)
indice_mejor = np.argmax(ventas_array)
indice_peor = np.argmin(ventas_array)

print(f"\n5. Análisis de ventas extremas:")
print(f"   Mejor venta: Venta {indice_mejor + 1} con ${mejor_venta:.2f}")
print(f"   Peor venta: Venta {indice_peor + 1} con ${peor_venta:.2f}")

print("\n----------------- RESUMEN FINAL ----------------")
print(f"Total de ventas analizadas: {cantidad_ventas}")
print(f"Promedio por venta: ${promedio_ventas:.2f}")
print(f"Total recaudado: ${total_recaudado:.2f}")
print(f"Ventas sobre el promedio: {len(valores_sobre_promedio)} de {cantidad_ventas}")
print(f"Diferencia entre mejor y peor venta: ${mejor_venta - peor_venta:.2f}")
```
